La meta es realizar un autómata (equivalente a una expresión regular) con el propósito de detectar las siguientes palabras:

- **"aud"**  
- **"regul"**

Estas palabras se podrán encontrar en un texto completamente en latín o en un texto dedicado al idioma latín. Por lo tanto, este analizador debe ser capaz de ubicar las cadenas **"aud"** y **"regul"**, e informar al usuario si dichos lexemas se encuentran en algún texto.

```python
import re # regex
```

```python
def regex_latin_word_identifier(latin_phrase):
    
    # encuentrar las cadenas 'aud' y 'regul'
    aud_findings = re.findall('(aud)+',latin_phrase) # automata numero uno
    regul_findings = re.findall('(regul)+',latin_phrase) # automata nuemro dos

    # tomar la cuenta

    aud_count = len(aud_findings)
    regul_count = len(regul_findings)

    if aud_count == 0:
        return print("No 'aud' found")
    elif regul_count == 0:
        return print("No 'regul' found")
    elif (aud_count == 0) and (regul_count == 0):
        return print("No correspodnece of the language.")
    elif (aud_count > 0) or (regul_count > 0):
        # indicar que uno u otra cadena fue ubicada exitosamente
        return "passed"
        

    
```

```python
# Frases Latinas

latin_phrases = [
    "Discipuli regulam audiunt, et magister regulas claras audit.",
    "Postea regulam novam audimus et meminimus.",
    "Regulae in schola scribuntur; discipuli diligenter audiunt.",
    "AUDITE, discipuli: haec est REGULA prima.",
    "Regulam, regulam! Discipuli tamen non audiunt.",
    "In foro regulam audivi; postea regulas multas audivimus.",
    "Si magister audiat, regula manet; si non audiat, regula cadit.",
    "Magister de regulis loquitur; discipulus vero: 'audio!'",
    "Regulae—si bene auditae sunt—facile servantur.",
    "Cum regulam audiam, regulas intellego; cum non audiam, erro.",
    "Discipuli legem audiunt, et magister praecepta explicat.",
    "Regulam servamus et meminimus; magister docet.",
    "audi regulam",
    "regula audit",
    "aud regul",
    "inauditus rex regulat populum.",
    "Regulā audiō.",
    "Regula123 audiunt456",
    "regulationes audiologist",
    ""
]
```

```python

# calar los casos de frases latinas 
count = 0

count += sum(
    1 for latin_x in latin_phrases
    if regex_latin_word_identifier(latin_x) == "passed"
)
```
